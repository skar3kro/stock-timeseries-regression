```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/kaggle/input/algo-trading-data-nifty-100-data-with-indicators/BHEL_minute.csv")

# Convert datetime column (adjust the column name if needed)
df['Datetime'] = pd.to_datetime(df['date'])

# Set as index
df = df.set_index('Datetime')

# Sort by time
df = df.sort_index()
df = df.drop('date', axis =1)

df.head()
```

```python
# Create OHLCV for 15-minute candles
df_15min = df.resample('15T').agg({
    'open': 'first',
    'high': 'max',
    'low': 'min',
    'close': 'last',
    'volume': 'sum'
}).dropna()

# Create OHLCV for 30-minute candles
df_30min = df.resample('30T').agg({
    'open': 'first',
    'high': 'max',
    'low': 'min',
    'close': 'last',
    'volume': 'sum'
}).dropna()

df_15min.head(), df_30min.head()
```

```python
# Create target for 15-min data: next close
df_15min['target'] = df_15min['close'].shift(-1)

# Drop last row where Target is NaN
df_15min = df_15min.dropna(subset=['target'])

# Same for 30-min data
df_30min['target'] = df_30min['close'].shift(-1)
df_30min = df_30min.dropna(subset=['target'])

df_15min.head()
```

```python
# ----- Add Lag Features -----

df_15min['lag_1'] = df_15min['close'].shift(1)
df_15min['lag_2'] = df_15min['close'].shift(2)
df_15min['lag_3'] = df_15min['close'].shift(3)
df_15min['lag_5'] = df_15min['close'].shift(5)
df_15min['lag_10'] = df_15min['close'].shift(10)

# Drop early NaNs from lag features
df_15min = df_15min.dropna()
```

```python
df_15min.head()
```

```python
import numpy as np

# ----- Feature Engineering on 15-min data -----

# Volatility inside candle
df_15min['spread'] = df_15min['high'] - df_15min['low']

# Candle body (direction + strength)
df_15min['body'] = df_15min['close'] - df_15min['open']

# 1-step return
df_15min['return_1'] = df_15min['close'].pct_change()

# Rolling features (use window=5 candles)
df_15min['ma_5'] = df_15min['close'].rolling(window=5).mean()
df_15min['vol_5'] = df_15min['close'].rolling(window=5).std()

# Drop rows with NaNs created by rolling/pct_change
df_15min = df_15min.dropna()

# ----- Feature matrix and target -----

feature_cols = [
    'open','high','low','close','volume',
    'spread','body','return_1','ma_5','vol_5',
    'lag_1','lag_2','lag_3','lag_5','lag_10'
]

X = df_15min[feature_cols]
y = df_15min['target']

split_index = int(len(df_15min) * 0.8)

X_train = X.iloc[:split_index]
X_test  = X.iloc[split_index:]
y_train = y.iloc[:split_index]
y_test  = y.iloc[split_index:]

```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

# Linear Regression
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)

y_pred_lr = lr.predict(X_test_scaled)

print("Linear Regression with Lags:")
print("MAE :", mean_absolute_error(y_test, y_pred_lr))
print("RMSE:", mean_squared_error(y_test, y_pred_lr, squared=False))
print("R2  :", r2_score(y_test, y_pred_lr))
```

```python
# ----- Feature engineering for 30-min data -----

# Basic candle features
df_30min['spread'] = df_30min['high'] - df_30min['low']
df_30min['body']   = df_30min['close'] - df_30min['open']
df_30min['return_1'] = df_30min['close'].pct_change()
df_30min['ma_5']     = df_30min['close'].rolling(5).mean()
df_30min['vol_5']    = df_30min['close'].rolling(5).std()

# Lag features (very important)
df_30min['lag_1']  = df_30min['close'].shift(1)
df_30min['lag_2']  = df_30min['close'].shift(2)
df_30min['lag_3']  = df_30min['close'].shift(3)
df_30min['lag_5']  = df_30min['close'].shift(5)
df_30min['lag_10'] = df_30min['close'].shift(10)

# Drop rows with NaNs from pct_change / rolling / lags
df_30min = df_30min.dropna()
```

```python
feature_cols_30 = [
    'open','high','low','close','volume',
    'spread','body','return_1','ma_5','vol_5',
    'lag_1','lag_2','lag_3','lag_5','lag_10'
]

X_30 = df_30min[feature_cols_30]
y_30 = df_30min['target']
```

```python
split_index_30 = int(len(df_30min) * 0.8)

X_train_30 = X_30.iloc[:split_index_30]
X_test_30  = X_30.iloc[split_index_30:]

y_train_30 = y_30.iloc[:split_index_30]
y_test_30  = y_30.iloc[split_index_30:]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Scale features
scaler_30 = StandardScaler()
X_train_30_scaled = scaler_30.fit_transform(X_train_30)
X_test_30_scaled  = scaler_30.transform(X_test_30)

# Train model
lr_30 = LinearRegression()
lr_30.fit(X_train_30_scaled, y_train_30)

# Predict
y_pred_30 = lr_30.predict(X_test_30_scaled)

print("30-min Linear Regression with Lags:")
print("MAE :", mean_absolute_error(y_test_30, y_pred_30))
print("RMSE:", mean_squared_error(y_test_30, y_pred_30, squared=False))
print("R2  :", r2_score(y_test_30, y_pred_30))
```
